```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from  sklearn.cluster import KMeans
import numpy as np
import itertools
from sklearn.metrics import silhouette_score
from sklearn.model_selection import  train_test_split
from  sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
```

```python
scaler = preprocessing.MinMaxScaler()
df = pd.read_csv('creditcard.csv')
classdf = df['Class']
print(classdf)
```

```python
newdf = df.drop(['Class'] , axis = 1)
coloumns_to_normalize = newdf.columns
newdf[coloumns_to_normalize] = scaler.fit_transform(newdf[coloumns_to_normalize])
newdf
```

```python
def best_k(data,max_k):
    means = []
    inertias =[]
    for k in range(1,max_k + 1):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        means.append(k)
        inertias.append(kmeans.inertia_)

    fig = plt.subplots(figsize = (10,5))
    plt.plot(means,inertias , 'o-')
    plt.xlabel('number of clusters')
    plt.ylabel('SSE')
    plt.grid(True)
    plt.show()

best_k(df,10)
```

```python
data = np.array(df_matrix)
labels = np.array(labels)

num_clusters = len(np.unique(labels))
for cluster in range(num_clusters):
    cluster_data = data[labels == cluster]

    means = np.mean(cluster_data, axis=0)
    variances = np.var(cluster_data, axis=0)
    np.set_printoptions(precision=2, suppress=True)
    print(f"\nخوشه {cluster + 1}: میانگین = {means}\n\n واریانس = {variances}")

```

```python
def fit_and_predict(data, n_clusters, random_state=123):
    
    km = KMeans(n_clusters=n_clusters, max_iter=300,
                random_state=random_state, init='k-means++').fit(data)
    y_pred = km.fit_predict(df_matrix)
    return y_pred, km


def plot_data_with_cluster(data, pred, n_clusters, km):
    
    _cycols = itertools.cycle('bgcmywk')
    _marker = itertools.cycle((',', '+', '.', 'o', '*'))
    plt.figure(figsize=(8, 8))
    for i in np.unique(pred):
        plt.scatter(
            data[pred == i, 0], data[pred == i, 1],
            s=50, c=next(_cycols),
            marker=next(_marker), edgecolor='black',
            alpha=.5, label='cluster {:d}'.format(i)
        )

    plt.scatter(
        km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
        s=300, marker='*',
        c='red', edgecolor='black', label='Centroids'
    )
    plt.legend(scatterpoints=1)
    plt.legend(loc='best')
    plt.show()
    return None
y_pred_elbow, elbow_model = fit_and_predict(df_matrix, 2)
plot_data_with_cluster(df_matrix, y_pred_elbow, 3,kmeans)
```
